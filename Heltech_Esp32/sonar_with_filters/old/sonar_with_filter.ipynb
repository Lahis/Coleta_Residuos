```python
# test dataframe visualization
dataFrame_12 = pd.read_csv("sonar_data_12.csv") 
dataFrame_18 = pd.read_csv("sonar_data_18.csv") 
dataFrame_22 = pd.read_csv("sonar_data_22.csv")  
dataFrame_29 = pd.read_csv("sonar_data_29.csv") 


dataFrame_29.head(22)
```

```python
# graph measures x time
plt.figure(figsize=(15,7))
ax = plt.gca()

dataFrame_12.plot(kind='line',x="time", y='raw', color = 'red', ax=ax)
dataFrame_12.plot(kind='line',x="time", y='mean', color = 'green', ax=ax)
dataFrame_12.plot(kind='line',x="time", y='median', color = 'black', ax=ax)
#dataFrame_12.plot(kind='line',x="time", y='mov_average', color = 'blue', ax=ax)


plt.xlabel('Time')
plt.ylabel('Measures (cm)')
plt.title('Measures x Time (Lixeira Cheia)')
plt.show()
```
